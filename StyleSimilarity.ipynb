```python
from sentence_transformers import SentenceTransformer, util

# Load the embedding model
model = SentenceTransformer('all-MiniLM-L6-v2')

# AI-generated style descriptions
input_styles = [
    "Short-sleeved, ruffled sleeves, keyhole neckline",
    "Short sleeve blouse with ruffled sleeves and keyhole neckline",
    "Short-sleeved, ruffled neckline and sleeves"
]

# Define standard styles from your vocabulary
style_vocabulary = {
    "T-shirt": ["Crew neck", "V-neck", "Short-sleeved", "Graphic", "Fitted", "Oversized"],
    "Blouse": ["Ruffled", "Puffed sleeves", "Silk", "Pleated", "Floral"],
    "Shirt": ["Button-down", "Collar", "Checked", "Striped", "Formal", "Casual"],
    "Tank Top": ["Sleeveless", "Ribbed", "Cropped", "Spaghetti strap"]
}

# Flatten style phrases for embedding
style_labels = [(cat, phrase) for cat, styles in style_vocabulary.items() for phrase in styles]
style_texts = [s[1] for s in style_labels]
style_embeddings = model.encode(style_texts, convert_to_tensor=True)

# Match each input to the best style
for query in input_styles:
    query_emb = model.encode(query, convert_to_tensor=True)
    scores = util.pytorch_cos_sim(query_emb, style_embeddings)[0]
    best_idx = scores.argmax().item()
    best_category, best_style = style_labels[best_idx]
    print(f"\nInput: {query}")
    print(f"→ Best Match: {best_category} / {best_style} (score: {scores[best_idx].item():.2f})")
```
